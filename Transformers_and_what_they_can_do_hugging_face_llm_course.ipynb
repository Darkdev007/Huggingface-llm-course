```python
#import the pipeline
from transformers import pipeline
```

```python
#Create the pipeline for specific tasks
#to understand the tone in a sentence
sentiment_pipe = pipeline("sentiment-analysis")
sentiment_pipe("I've been waiting for hugging face course all my life")
```

```python
sentiment_pipe(["I've been waiting for hugging face course all my life", "I hate my life"])
```

```python
#Zero shot classification
zero_short_pipeline = pipeline("zero-shot-classification")
zero_short_pipeline(
    "This is a course about transformers model",
    candidate_labels = ["politics", "education", "business"]
)
```

```python
#Text generation
text_generation_pipeline = pipeline("text-generation",  model = "HuggingFaceTB/SmolLM2-360M")
# text_generation_pipeline("The best way to protect yourself is to")
```

```python
text_generation_pipeline("Today will be a great day because",
                         max_length = 50,
                         num_return_sequences = 2)
```

```python
text_generation_pipeline("The best way to protect yourself is to",
                         max_length=15,
                         num_return_sequences=2)
```

```python
mask_filling_pipeline = pipeline("fill-mask")
```

```python
mask_filling_pipeline("This is going to be a <mask> day cause rain is falling", top_k=3)
```

```python
ner_pipeline = pipeline("ner", grouped_entities=True)
```

```python
ner_pipeline("The boy named tobi comes from nigeria and works as an ai engineer earning $20/hr for his first entry role,")
```

```python
question_answering_pipeline = pipeline("question-answering")
question_answering_pipeline(
    question = "How much do I earn?",
    context = "The boy named tobi comes from nigeria and works as an ai engineer earning $20/hr for his first entry role" )
```

```python
translation_pipeline = pipeline("translation", model="Helsinki-NLP/opus-mt-fr-en")
translation_pipeline("Ce cours est produit par Hugging Face.")
```

```python
from transformers import pipeline

image_classifier = pipeline(
    task = "image-classification", model="google/vit-base-patch16-224"
)
```
